```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

# 2. 
X, Y = np.meshgrid(x, y)

# 3.  f(x, y)
Z = X**2 + X*Y + Y**2

# 4. 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 5. 
ax.plot_surface(X, Y, Z)

# 6. 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

# 7. 
plt.show()
```
